```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

### 1.2 Изучите стоимости недвижимости



```python

plt.figure(figsize=(5,5))
plt.hist(x=df['price'])
plt.title("Стоимость недвижимости")
plt.xlabel('Mln USD')
plt.ylabel("Количество домов");
```

<span style = color:green> Вывод<br>
Большая часть домов стоимостью до 1 mln долларов </span> 

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(5,5))
plt.hist(x=df['sqft_living'], color='green', edgecolor='black')
plt.title("Жилая квадратура")
plt.xlabel('foot')
plt.ylabel("Количество домов");
```

<span style = color:green> Вывод<br>
Большая часть домов квадратурой около 3000 футов, около 7500 домов квадратурой около 2000 футов, затем идет снижение </span> 

### 2.1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(5,5))
plt.hist(x=df['yr_built'], color='green', edgecolor='black')
plt.title("Года постройки")
plt.xlabel('year')
plt.ylabel("Количество домов");
```

<span style = color:green> Вывод<br>
Прослеживается нормальное распределение по количеству построенных домов с 1950 по 2020 годы, значительно меньше домов построено с 1900 по 1950 годы (хотя тоже прослеживается нормальное распределение в эти годы)  </span> 

## Условие 2: 2 задача
### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график<br>
Сделайте выводы

```python
data = df['waterfront'].value_counts()
names = data.index
values = data.values
plt.pie(values, autopct='%.1f%%', labels=names)
plt.title('Вид на набережную');
```

<span style = color:green> Вывод<br>
Только 0.8% домов имеют вид на набережную </span> 

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
names = data.index
values = data.values
plt.figure(figsize=(4,5))
plt.bar(names,values)
plt.title('Количество этажей');
```

<span style = color:green> Вывод<br>
Больше всего одноэтажных, затем двухэтажных домов</span> 

### 2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
names = data.index
values = data.values
plt.figure(figsize=(4,5))
plt.bar(names,values)
plt.title('Состояние домов');
```

<span style = color:green> Вывод<br>
Больше всего домов состояния 3, затем 4 и 5</span> 

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix,1)
corr_matrix[np.abs(corr_matrix < 0.3)] =0 
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm');
```

<span style = color:green> Вывод<br>
Как видно выше показатели цены наиболее сильно связаны с грейдом недвижимости, общей площади, жилойплощади и в среднем от количества ванных комнат</span> 

```python
sns.relplot(df, x='grade', y='price', kind='line');
```

<span style = color:green> Вывод<br>
Как видно на графике цена меняется в зависимости от грейда,  причем после 10 грейда цена резко возврастает</span>

```python
sns.relplot(df, x='sqft_living', y='price');
```

<span style = color:green> Вывод<br>
Как видно на графике цена меняется в зависимости от жилой площади (имеются отдельные выбросы)</span>

```python
sns.relplot(df, x='sqft_above', y='price', hue="sqft_above");
```

<span style = color:green> Вывод<br>
Аналогичная картина наблюдается зависимости цены от общей площади</span>
